```python
from openai import OpenAI

client = OpenAI(
    base_url="http://localhost:11434/v1",
    api_key="not-needed"
)
# Using streaming for better User Experience
stream = client.chat.completions.create(
    model="gemma4:e2b",
    messages=[
        {"role": "system", "content": "You are a helpful coding assistant."},
        {"role": "user", "content": "Write a Python function that merges two sorted lists."}
    ],
    temperature=0.7,
    stream=True,
)
for chunk in stream:
    if chunk.choices[0].delta.content:
        print(chunk.choices[0].delta.content, end="", flush=True)
```

```python
import base64

def encode_image(image_path):
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode('utf-8')
base64_image = encode_image("petrol.png")
response = client.chat.completions.create(
    model="gemma4:e2b   ",
    messages=[
        {
            "role": "user",
            "content": [
                {"type": "text", "text": "Analyze this image and get me the details."},
                {
                    "type": "image_url",
                    "image_url": {"url": f"data:image/jpeg;base64,{base64_image}"}
                },
            ],
        }
    ],
)
print(response.choices[0].message.content)
```

```python
import base64

def encode_audio(audio_path):
    with open(audio_path, "rb") as audio_file:
        return base64.b64encode(audio_file.read()).decode('utf-8')

audio_base64 = encode_audio("harvard.wav")

response = client.chat.completions.create(
    model="gemma4:e2b",
    messages=[
        {
            "role": "user",
            "content": [
                {"type": "text", "text": "Transcribe this audio exactly as spoken."},
                {
                    "type": "input_audio",
                    "input_audio": {
                        "data": audio_base64,
                        "format": "wav"  # Ensure format matches your file (wav, mp3, etc.)
                    }
                },
            ],
        }
    ],
)

print(response.choices[0].message.content)
```
