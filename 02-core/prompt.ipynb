## Prompt Template

```python
from langchain_core.prompts import \
( SystemMessagePromptTemplate, 
 HumanMessagePromptTemplate,
  ChatPromptTemplate)
```

```python
system = SystemMessagePromptTemplate.from_template\
    ("You are a {school} teacher, You answer in short sentence"
)
question = HumanMessagePromptTemplate.from_template\
    ("tell me about the {topics} in {points} points")
```

```python
system
```

```python
question
```

```python
question.format(topics="sun", points=5)
```

```python
system.format(school='primary')
```

```python
messages =[system, question]
```

```python
messages
```

```python
template = ChatPromptTemplate(messages=messages)
```

```python
template
```

```python
question = template.invoke({'school':'phd', 'topics': 'earth', 'points': 3})
```

```python
question
```

```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model='gpt-4o-mini')
response = llm.invoke(question)
```
