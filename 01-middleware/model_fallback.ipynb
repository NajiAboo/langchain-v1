## Model fallback

*  Automatically fallback to alternative models when the primary model fails
* Building resilent agents that handle model outages.
* Cost optimization by falling back to cheaper models 
* Provide redundancy across OpenAI , Anthropic etc

```python
from langchain.agents import create_agent
from langchain.agents.middleware import ModelFallbackMiddleware
```

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
agent = create_agent(model="gpt-4o-mini",
                     tools=[],
                     )
```

```python
agent.invoke({
    "messages": [{'role': 'user', 'content': 'hi'}]
})
```

```python
agent = create_agent(model="gpt-4o-mini",
                     tools=[],
                     middleware=[
                         ModelFallbackMiddleware(
                             'gpt-4.1',
                             'claude-3.5-sonnet'
                         ),
                     ],
                     )
```

```python
agent.invoke({
    "messages": [{
        'role': 'user', 'content': "hi"
    }]
})
```
